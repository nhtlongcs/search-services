```python
import os
from pathlib import Path
import pandas as pd
```

```python
data_folder_path = Path('~/lsc23/meta/').expanduser()
metadata_path = data_folder_path / 'lsc22_metadata.csv'
visual_concept_path = data_folder_path / 'lsc22_visual_concepts.csv'
gps_data_path = data_folder_path / 'vaisl_gps.csv'
```

```python
metadata = pd.read_csv(metadata_path, low_memory=False, index_col=0)
vc_data = pd.read_csv(visual_concept_path, low_memory=False, index_col=0)
gps_data = pd.read_csv(gps_data_path, low_memory=False, index_col=0)
```

```python
metadata['ImageID'].fillna('', inplace=True)
metadata['index'] = metadata['ImageID'].apply(lambda x: x.split('.')[0])
metadata.head()
```

```python
vc_data['index'] = vc_data['ImageID'].apply(lambda x: x.split('.')[0])
vc_data.head()
```

```python
gps_data['index'] = gps_data.index
gps_data['index'] = gps_data['index'].apply(lambda x: x.split('.')[0])
gps_data.reset_index(drop=True, inplace=True)
gps_data.head()
```

```python
df = gps_data.merge(vc_data, on='index')
df.head()
```

```python
df.to_csv('lsc23.csv', index=False)
len(set(gps_data['index']).intersection(set(vc_data['index'])))
```
